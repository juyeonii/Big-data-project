```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rc
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
#한글 폰트(맑은 고딕) 적용
matplotlib.rc("font", family = "NanumGothic")
matplotlib.rc("axes", unicode_minus = False)
df=pd.read_csv("/home/piai/바탕화면/data.csv",encoding="cp949")
```

```python
df[["TempAvg"]].describe()
```

```python
df.plot(kind ="box",
           figsize =(50,10), subplots = True, sharex = False, sharey = False)
```

```python
df_duplicate=df.drop_duplicates("TempAvg")
b=pd.DataFrame(df_duplicate["TempAvg"])
a=pd.DataFrame(df.groupby("TempAvg").agg('sum')["Price"])
TempAvg=a.merge(b,on="TempAvg",how="left")
TempAvg
#온도별 매출 
plt.figure(figsize=(10,5))
plt.title("온도별 매출 변화")
plt.plot(TempAvg['TempAvg'],TempAvg["Price"], "-", color='grey')
plt.grid()
plt.show()
```

```python
df_duplicate=df.drop_duplicates("HumidAvg")
b=pd.DataFrame(df_duplicate["HumidAvg"])
a=pd.DataFrame(df.groupby("HumidAvg").agg('sum')["Price"])
HumidAvg=a.merge(b,on="HumidAvg",how="left")
HumidAvg
#습도별 매출 
plt.figure(figsize=(10,5))
plt.title("습도별 매출 변화")
plt.plot(HumidAvg['HumidAvg'], HumidAvg["Price"], "-", color='grey')
plt.grid()
plt.show()
```
