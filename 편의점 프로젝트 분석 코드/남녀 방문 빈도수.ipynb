```python
df1 = df.drop_duplicates(subset= ["idUser", "SalesDate"])
df1.head()
```

```python
data_gen = df1.groupby([df1["idUser"],df1['SalesDate'].str[:7],df1["Gender"]]).size().reset_index()
data_gen = pd.DataFrame(data_gen)
data_gen
```

```python
data_gen = data_gen.rename(columns={0:'방문수'})
data_gen
```

```python
data_mean = data_gen.groupby([data_gen["idUser"],data_gen["Gender"]])["방문수"].agg("mean").reset_index()
data_mean
```

```python
data_2 = data_mean.query("방문수>20")
data_2
```

```python
data_2 = data_2.groupby([data_2["Gender"]]).size().reset_index()
data_2
```

```python
data_3 = data_mean.query("10<방문수<=20")
data_3 = data_3.groupby([data_3["Gender"]]).size().reset_index()
data_3
```
